```python
import pandas as pd

df = pd.read_csv("Online_Retail.csv")

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]

cust_id = float(input("Enter Customer ID: "))

result = df[df["CustomerID"] == cust_id]

if len(result) > 0:

    spent = result["TotalAmount"].sum()

    print("Total Spent =", spent)

    print("Expected Purchase =", spent * 1.1)

else:

    print("Customer Not Found")
```

```python
import pandas as pd

df = pd.read_csv("Online_Retail.csv")

df["TotalAmount"] = df["Quantity"] * df["UnitPrice"]

cust_id = float(input("Enter Customer ID: "))

result = df[df["CustomerID"] == cust_id]

if len(result) > 0:

    spent = result["TotalAmount"].sum()

    if spent >= 5000:
        segment = "Premium Customer"
    elif spent >= 1000:
        segment = "Regular Customer"
    else:
        segment = "Low Value Customer"

    print("Segment =", segment)
    print("Total Spent =", spent)
    print("Expected Purchase =", spent * 1.1)

else:
    print("Customer Not Found")
```
